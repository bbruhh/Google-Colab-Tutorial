```
# Upload Files & Data - Google Colab Tutorial

######################################################

from google.colab import files

uploaded = files.upload()

print(uploaded)

uploaded = uploaded["NVDA.csv"]

print(uploaded)

for item in uploaded:
    print(item)
    
######################################################

print(type(uploaded))

uploaded = str(uploaded)

print(type(uploaded))

uploaded = uploaded.split("\\n")

print(uploaded)

for item in uploaded:
    print(item)
    
new_list = []

for item in uploaded:
    new_list.append(item.split(","))
    
print(new_list)

for item in new_list:
    print(item)
```
